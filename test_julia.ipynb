```julia
using Plots
n, d = 20, 3
x = range(-, ; length=n)
X = x .^ (0:d)'
y = sin.(x) + 0.3randn(n)
b = X\y
f(x) = evalpoly(x, b)  # 3
xs = range((extrema(x) .* 1.1)...; length=1000)
plot(xs, f; label="degree-$d polynomial")
plot!(xs, sin; label="sin(x)", ls=:dash)
scatter!(x, y; label="sample", color=:red)
plot!(; legend=:topleft)
```

```julia
function oreore(x, b)
    x^0*b[1] + x^1*b[2] + x^2*b[3] + x^3*b[4] 
end

oreore_f(x) = oreore(x, b)
plot(xs, oreore_f; label="degree-$d polynomial")
```
