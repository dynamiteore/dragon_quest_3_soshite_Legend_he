## 分類 Classification  
### 2次元入力2クラス分類の場合  



```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(seed = 1)
N = 100
K = 3
T3 = np.zeros((N,3), dtype=np.uint8)
T2 = np.zeros((N,2), dtype=np.uint8)
X = np.zeros((N,2))
X_range0 = [-3, 3]
X_range1 = [-3, 3]
Mu = np.array([[-.5, -.5], [.5, 1.0], [1, -.5]]) # 分布の中心
Sig = np.array([[.7, .7], [.8, .3], [.3, .8]])  # 分布の分散
Pi = np.array([0.4, 0.8, 1])
for n in range(N):
    # wk:0～1 までの一様分布乱数
    # クラス0　　wk < Pi[0]
    # クラス1　　wk >= Pi[0] && wk < Pi[1]
    # クラス2　　wk >= Pi[1] && wk < Pi[2]
    wk = np.random.rand()
    for k in range(K):
        if wk < Pi[k]:
            T3[n, k] = 1
            break
    for k in range(2):
        X[n, k] = (np.random.randn() * Sig[T3[n, :] == 1, k] + Mu[T3[n, :] == 1, k])

# 1-of-K符号化法ｓ
T2[:, 0] = T3[:, 0]
T2[:, 1] = T3[:, 1] | T3[:, 2]
```

```python
# データ表示
def show_data2(x, t):
    wk, K = t.shape
    c = [[.5,  .5, .5], [1, 1, 1], [0, 0, 0]]
    for k in range(K):
        plt.plot(x[t[:, k] == 1, 0], x[t[:, k] == 1, 1], ls = 'none', mec='black', marker='o', c=c[k], alpha=0.8)
    plt.grid(True)

plt.figure(figsize=(7.5, 3))
plt.subplot(1,2,1)
show_data2(X, T2)
plt.xlim(X_range0)
plt.ylim(X_range1)

plt.subplot(1,2,2)
show_data2(X, T3)
plt.xlim(X_range0)
plt.ylim(X_range1)

plt.show()
```

ロジスティック回帰  
$$
y = σ(a) = \frac{1}{1 + \exp(-a)}
$$  

$$
a = w_0x_0 + w_1x_1 + w_2
$$

```python
# ロジスティック回帰モデル
def logistic2(x0, x1, w):
    y = 1 / (1 + np.exp(-(w[0]*x0 + w[1]*x1 + w[2])))
    return y
```

## 平均交差エントロピー誤差関数  
2クラス分類なので前回と同じ式  
$$
E(\boldsymbol{w}) 
= - \frac{1}{N} \log P(\boldsymbol{T}|\boldsymbol{X}) 
= - \frac{1}{N} \sum_{n=0}^{N-1}\{t_n \log y_n + (1-t_n) \log(1-y_n)\}
$$

```python
# 交差エントロピー誤差
def cee_logistic2(w, x, t):
    X_n = x.shape[0]
    y = logistic2(x[:, 0], x[:, 1], w)
    cee = 0
    for n in  range(len(y)):
        cee = cee - (t[n, 0] * np.log(y[n]) + (1 - t[n, 0]) * np.log(1 - y[n]))
    cee = cee / X_n
    return cee
```

```python
# 交差エントロピー誤差微分
def dcee_logistic2(w, x, t):
    X_n = x.shape[0]
    y = logistic2(x[:, 0], x[:, 1], w)
    dcee = np.zeros(3)
    for n in range(len(y)):
        dcee[0] = dcee[0] + (y[n] - t[n, 0]) * x[n, 0]
        dcee[1] = dcee[1] + (y[n] - t[n, 0]) * x[n, 1]
        dcee[2] = dcee[2] + (y[n] - t[n, 0])
    dcee = dcee / X_n
    return dcee

W = [-1,-1,-1]
dcee_logistic2(W, X, T2)
```

```python
# 平均交差エントロー誤差が最小になるようにロジスティック回帰モデルのパラメータを求める
from scipy.optimize import minimize

def fit_logistic2(w_init, x, t):
    res = minimize(cee_logistic2, w_init, args=(x, t), jac=dcee_logistic2, method="CG")
    return res.x

fig_1 = plt.figure( figsize=(10, 4.5))
plt.subplots_adjust(wspace=0.5)

Ax = plt.subplot(1,2,1, projection='3d')
W_init = [-1,0,0]
W = fit_logistic2(W_init, X, T2)
print("w0 = {0:.2f}, w1 = {1:.2f}, w2 = {2:.2f}".format(W[0], W[1], W[2]))
show3d_logistic2(Ax, W)
show_data2_3d(Ax, X, T2)
cee = cee_logistic2(W, X, T2)
print("CEE = {0:.2f}".format(cee))

#Ax = fig_1.add_subplot(111)
Ax = plt.subplot(1,2,2)
show_data2(X, T2)
show_contour_logistic2(W)


plt.show()
```

# 2次元入力2クラス分類おさらい
・入力　パラメータが3つになる$w0, w1, w2$  
$$ a = w_0x_0 + w_1x_1 + w_2 $$  
・出力　t=0である確率近似 $P(t = 0|x)$、t=1  $P(t = 1|x)$  
・シグモイド関数に入力の式を当てはめてロジスティック回帰モデルを作る  
・平均交差エントロピー誤差関数は1次元2クラス分類と同じ式が使える  
・交差エントロピー微分も同様に式を使う  
・勾配法で平均交差エントロピー誤差が最小になるように求める  
・3Dグラフのy=0.5等高線が決定境界になる
